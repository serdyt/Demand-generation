```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/K2 2019-11 dead no id_hour.csv', sep=';')
```

```python
df['kost'] = [s.replace(',', '.') for s in df['förd kost'].to_list()]
df['kost'] = df['kost'].astype(float)

df['km'] = [s.replace(',', '.') for s in df['p km'].to_list()]
df['km'] = df['km'].astype(float)

df['kost_per_km'] = df.kost / df.km
```

```python
df.head()
```

```python
df = df[df.kost_per_km != 0]
df = df[df.kost_per_km != np.inf]
df = df[df.kost_per_km != np.NINF]
df = df[df.kost_per_km != np.NAN]
df = df[df.kost_per_km > 5]
df = df[df.kost_per_km < 200]
df = df.dropna(subset=['kost_per_km'])
```
